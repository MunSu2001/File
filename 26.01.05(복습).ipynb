```python
class Student:
    def __init__(self, name):
        self._name = name 

    @property  # 읽기 전용 속성 만들기 
    def name(self):
        return self._name

    @name.setter   # setter = instance 값을 변경하기 위해 필요
    def name(self, value): # 없어도 코드는 돌아가지만 추후에 변경 불가 
        self._name = value

    @name.getter  # getter = instance 값을 읽기 위해 필요 하지만 property에 포함되어 있어 없어도 정상 작동
    def name(self): # 지금은 그냥 눈으로 확인하기 위해 작성
        return self._name
```

```python
a = Student('원빈')
```

```python
a._name # 인스턴스 속성에 접근, ._name은 일반 변수 
        # => 단순히 self._name = name으로 저장된 값을 꺼내는것
```

```python
class Student:
    def __init__(self, name, math):
        self._name = name 
        self._math = math

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        print('setter')
        self._name = value

    @name.getter
    def name(self):
        print("getter")
        return self._name

    @property
    def math(self):
        return self._name

    @math.setter
    def math(self, value):
        print('setter -> {}'.format(value))
        if 0<= value <= 100:
            self._math = value
        else:
            raise ValueError("점수 똑바로 넣어!!")

    @math.getter
    def math(self):
        print("getter")
        return self._math

    def __str__(self):  
        return self._name
b = Student('엔코아', 90)
```

```python
try:
    b.math = 101
except Exception as e:
    print(e)
```

```python
tmp = {'a':99, 'b':99, 'c':65 , 'd':99}

sorted(tmp.items(), key=lambda x: (x[0]))
# x[0]은 키 값 의미 
# 즉 a b c d를 오름차순으로 나열 
# * default 값은 오름차순 
# 만약 내림 차순으로 하고 싶으면 reverse = False 입력
```

```python
sorted(tmp.items(), key= lambda x:(-x[1]))
# -x[1]은 value 값에 -를 붙인다 즉 99 99 65 99 -> -99 -99 -65 -99 로 변환
# 변환 후 오름차순으로 나열시 -99 -99 -99 -65로 정렬 
# 정렬 기준으로는 움수 값이 사용되지만 실제 반환은 원래 값 value 그대로 반환
# 내림차순으로 보일순 있으나 내부적인 회로는 오름차순임
```

```python
sorted(tmp.items(), key=lambda x: (x[1], x[0])) # 다중 기준
# 먼저 앞에 있는 x[1]을 실행 후 x[0] 실행
# 즉 value 값이 동일한 key 값을 오름차순으로 나열 
```
