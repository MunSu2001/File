```python
df[df.연봉.apply(lambda x : x != "") & df.HEIGHT.apply(lambda x : len(x) > 2)] # 데이터프레임 안에서는 비트 연산자 사용 
```

```python
df2 = df[df.연봉.apply(lambda x : x != "") & df.HEIGHT.apply(lambda x :  len(x) > 2)].copy()
```

```python
df2.shape
```

```python
df2['BMI'] = df2.WEIGHT.apply(lambda x : int(p.findall(x)[0])) /  \
    (df2.HEIGHT.apply(lambda x : int(p.findall(x)[0])) / 100 ) ** 2
```

```python
df2.HEIGHT = df2.HEIGHT.apply(lambda x : x.replace("cm", "")).astype(int)
df2.WEIGHT = df2.WEIGHT.apply(lambda x : x.replace("kg", "")).astype(int)
```

```python
def temp(x):
    if x [-2:] == '만원':
        return int(x[:-2])
    else:
        return int(x[:-2]) * 1450 /10000
```

```python
df2.연봉 = df2.연봉.apply(temp)
```

```python
trans_dict = {'선수명' : 'NAME',
'등번호' : 'NUMBER',
'생년월일' : 'BIRTH_DAY',
'포지션' : 'POSITION',
'경력' : 'CAREER', 
'입단 계약금' : 'SIGNING_BONUS',
'연봉' : 'SALARY',
'지명순위' : 'DRAFT_PICK', 
'입단년도' : 'JOIN_YEAR',
'Team' : 'TEAM', 
'id' : 'ID'}
```

```python
df2.rename(columns = trans_dict, inplace=True)
```
